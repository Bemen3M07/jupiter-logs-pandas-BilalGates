De la mateixa manera que a l'exercici 5 caldrà investigar una llibreria per a realitzar gràfics de dades visuals i elaborar com a mínim el codi necessari per presentar de forma que tinguin sentit amb l'àmbit del problema.

Es pot utilitzar la llibreria python matploit

```python
import pandas as pd
import matplotlib.pyplot as plt

# Llegir dades des del fitxer CSV
df = pd.read_csv("dades.csv")

# Comprovar les columnes existents
print(df.columns)

# Afegir columna amb la mitjana (ajusta segons el teu fitxer)
# Suposant que les columnes de notes comencen des de la segona columna fins al final
df["Mitjana"] = df.iloc[:, 1:].mean(axis=1)

# Generar un histograma de les mitjanes
df["Mitjana"].hist()
plt.title("Distribució de Mitjanes")
plt.xlabel("Mitjana de Notes")
plt.ylabel("Freqüència")
plt.show()
```
